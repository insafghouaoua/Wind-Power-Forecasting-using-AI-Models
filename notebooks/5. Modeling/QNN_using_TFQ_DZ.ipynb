#Model

**Model QNN avec TFQ (Tensorflow Quantique)**

1. Imports des bibliothèques

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import tensorflow_quantum as tfq
import cirq
import sympy
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from tqdm import tqdm
import joblib
```

2. Chargement des données

```python
train = pd.read_csv('/content/donnees_train.csv')
val = pd.read_csv('/content/donnees_val.csv')
test = pd.read_csv('/content/donnees_test.csv')
```

3. Définition des colonnes

```python
target_column = "PuissanceActive (kW)"
quantum_features = ['VitesseVent (m/s)', 'DirectionVent (°)', 'Temperature (°C)', 'Saison']
classical_features = [
    'Annee', 'Mois', 'Jour', 'Heure',
    'JourNuit', 'deviation', 'deviation_percentage', 'usability'
]
all_features = quantum_features + classical_features
```

4. Préparation des données d'entraînement et validation

```python
X_train, y_train = train[all_features], train[target_column]
X_val, y_val = val[all_features], val[target_column]
```

5. Filtrage selon la vitesse du vent

```python
def filter_by_wind(X, y):
    mask = (X['VitesseVent (m/s)'] > 3) & (X['VitesseVent (m/s)'] <= 25)
    return X[mask], y[mask]

X_train, y_train = filter_by_wind(X_train, y_train)
X_val, y_val = filter_by_wind(X_val, y_val)
```

6. Normalisation des données

```python
scaler_X = MinMaxScaler()
X_train_scaled = pd.DataFrame(scaler_X.fit_transform(X_train), columns=all_features)
X_val_scaled = pd.DataFrame(scaler_X.transform(X_val), columns=all_features)

scaler_y = MinMaxScaler()
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_val_scaled = scaler_y.transform(y_val.values.reshape(-1, 1)).flatten()
```

7. Définition du circuit quantique

```python
n_qubits = len(quantum_features)
qubits = [cirq.GridQubit(0, i) for i in range(n_qubits)]

def create_quantum_circuit():
    params = sympy.symbols(f'theta0:{n_qubits}')
    circuit = cirq.Circuit()
    for i, q in enumerate(qubits):
        circuit.append(cirq.ry(params[i])(q))
    for i in range(n_qubits - 1):
        circuit.append(cirq.CNOT(qubits[i], qubits[i + 1]))
    observables = [cirq.Z(q) for q in qubits]
    return circuit, params, observables

model_circuit, params, observables = create_quantum_circuit()

# Affichage du circuit
print("Structure du circuit quantique:")
print(model_circuit)
```

8. Encodage quantique des données

```python
def generate_quantum_data(X_data):
    data_circuits = []
    for _, row in tqdm(X_data.iterrows(), total=len(X_data), desc="Génération des circuits"):
        features = np.pi * np.tanh(row[quantum_features].values)
        resolver = dict(zip(params, features))
        resolved = cirq.resolve_parameters(model_circuit, resolver)
        data_circuits.append(resolved)
    return tfq.convert_to_tensor(data_circuits)

train_circuits = generate_quantum_data(X_train_scaled)
val_circuits = generate_quantum_data(X_val_scaled)
```

9. Données classiques

```python
X_train_classical = X_train_scaled[classical_features].values
X_val_classical = X_val_scaled[classical_features].values
```

10. Définition du modèle hybride

```python
def create_hybrid_model():
    input_circuit = tf.keras.Input(shape=(), dtype=tf.string, name='quantum_input')
    quantum_output = tfq.layers.PQC(model_circuit, observables)(input_circuit)

    input_classical = tf.keras.Input(shape=(len(classical_features),), name='classical_input')
    x_classical = tf.keras.layers.Dense(16, activation='relu')(input_classical)

    x = tf.keras.layers.concatenate([quantum_output, x_classical])
    x = tf.keras.layers.Dense(16, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
    x = tf.keras.layers.Dropout(0.3)(x)
    x = tf.keras.layers.Dense(8, activation='relu')(x)
    x = tf.keras.layers.Dropout(0.2)(x)
    output = tf.keras.layers.Dense(1)(x)

    return tf.keras.Model(inputs=[input_circuit, input_classical], outputs=output)

def weighted_mse(y_true, y_pred):
    weight = tf.math.reciprocal(tf.math.abs(y_true) + 1e-2)
    return tf.reduce_mean(weight * tf.square(y_true - y_pred))

hybrid_model = create_hybrid_model()
hybrid_model.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss=weighted_mse, metrics=['mae'])
```

11. Entraînement du modèle

```python
early_stop = tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True, verbose=1)
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(patience=5, factor=0.5, min_lr=1e-6, verbose=1)

history = hybrid_model.fit(
    [train_circuits, X_train_classical], y_train_scaled,
    validation_data=([val_circuits, X_val_classical], y_val_scaled),
    epochs=50,
    batch_size=16,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)
```

12. Préparation des données de test

```python
test.dropna(inplace=True)
X_test, y_test = test[all_features], test[target_column].reset_index(drop=True)
X_test_masked, y_test_masked = filter_by_wind(X_test, y_test)
X_test_scaled = pd.DataFrame(scaler_X.transform(X_test_masked), columns=all_features)

test_circuits = generate_quantum_data(X_test_scaled)
X_test_classical = X_test_scaled[classical_features].values
```

13. Prédiction sur les données de test

```python
y_pred_scaled = hybrid_model.predict([test_circuits, X_test_classical])
y_pred = scaler_y.inverse_transform(y_pred_scaled).flatten()

y_pred_final = np.zeros(len(X_test))
mask = (X_test['VitesseVent (m/s)'] > 3) & (X_test['VitesseVent (m/s)'] <= 25)
y_pred_final[mask] = y_pred
```

14. Évaluation du modèle

Avec mask

```python
# Masquage : données où le modèle a été réellement appliqué
y_test_masked = y_test[mask].reset_index(drop=True) / 1000  # en MW
y_pred_masked = y_pred / 1000

# Calcul des métriques sur les données masquées (production réelle)
mse = mean_squared_error(y_test_masked, y_pred_masked)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test_masked, y_pred_masked)
r2 = r2_score(y_test_masked, y_pred_masked)
mape = np.mean(np.abs((y_test_masked - y_pred_masked) / y_test_masked.clip(lower=1e-8))) * 100

range_y = np.max(y_test_masked) - np.min(y_test_masked)
nrmse_range = (rmse / range_y) * 100 if range_y > 0 else float('inf')
nmse_range = (mse / (range_y ** 2)) * 100 if range_y > 0 else float('inf')
nmae = (mae / range_y) * 100 if range_y > 0 else float('inf')

denom = (np.abs(y_test_masked) + np.abs(y_pred_masked)).clip(lower=1e-8)
smape = np.mean(2.0 * np.abs(y_pred_masked - y_test_masked) / denom) * 100

# Affichage des résultats
print("Évaluation sur les données avec production (3 < Vent ≤ 25 m/s) :")
print(f" - MSE    : {mse:.4f} MW²")
print(f" - RMSE   : {rmse:.4f} MW")
print(f" - MAE    : {mae:.4f} MW")
print(f" - R²     : {r2:.4f}")
print(f" - MAPE   : {mape:.4f}%")
print(f" - SMAPE  : {smape:.4f}%")
print(f" - NMSE   : {nmse_range:.4f}%")
print(f" - NRMSE  : {nrmse_range:.4f}%")
print(f" - NMAE   : {nmae:.4f}%")
```

Sur l'enssemble des données

```python
# Conversion en MW
y_test_mw = y_test / 1000
y_pred_mw = y_pred_final / 1000

# Calcul des métriques en MW
mse = mean_squared_error(y_test_mw, y_pred_mw)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test_mw, y_pred_mw)
r2 = r2_score(y_test_mw, y_pred_mw)
mape = np.mean(np.abs((y_test_mw - y_pred_mw) / y_test_mw.clip(lower=1e-8))) * 100

range_y = np.max(y_test_mw) - np.min(y_test_mw)
nrmse_range = (rmse / range_y) * 100 if range_y > 0 else float('inf')
nmse_range = (mse / (range_y ** 2)) * 100 if range_y > 0 else float('inf')
nmae = (mae / range_y) * 100 if range_y > 0 else float('inf')

denom = (np.abs(y_test_mw) + np.abs(y_pred_mw)).clip(lower=1e-8)
smape = np.mean(2.0 * np.abs(y_pred_mw - y_test_mw) / denom) * 100

# Affichage des résultats
print("Évaluation sur le jeu de test (en MW) :")
print(f" - MSE    : {mse:.4f} MW²")
print(f" - RMSE   : {rmse:.4f} MW")
print(f" - MAE    : {mae:.4f} MW")
print(f" - R2     : {r2:.4f}")
print(f" - MAPE   : {mape:.4f}%")
print(f" - SMAPE  : {smape:.4f}%")
print(f" - NMSE   : {nmse_range:.4f}%")
print(f" - NRMSE  : {nrmse_range:.4f}%")
print(f" - NMAE   : {nmae:.4f}%")
```

17. Évaluation par Plages de Production

Avec mask

```python
# Conversion en MW
y_test_mw = y_test / 1000
y_pred_final_mw = y_pred_final / 1000

# Masque des prédictions
mask_vent = (X_test['VitesseVent (m/s)'] > 3) & (X_test['VitesseVent (m/s)'] <= 25)

def eval_par_plage(y_true, y_pred, plage_name, mask_plage):
    # On applique aussi le masque du vent pour ne prendre que les vraies prédictions
    maskall = mask_plage & mask_vent
    y_true_plage = y_true[maskall]
    y_pred_plage = y_pred[maskall]

    if len(y_true_plage) == 0:
        print(f"\nAucune donnée dans la plage '{plage_name}' avec conditions de vent valides.")
        return

    range_y_mw = np.max(y_true_plage) - np.min(y_true_plage)

    mse_p = mean_squared_error(y_true_plage, y_pred_plage)
    rmse_p = np.sqrt(mse_p)
    mae_p = mean_absolute_error(y_true_plage, y_pred_plage)
    r2_p = r2_score(y_true_plage, y_pred_plage)
    mape_p = np.mean(np.abs((y_true_plage - y_pred_plage) / y_true_plage.clip(lower=1e-8))) * 100
    nmae_p = (mae_p / range_y_mw) * 100 if range_y_mw > 0 else float('inf')
    smape_p = np.mean(2.0 * np.abs(y_pred_plage - y_true_plage) /
                      (np.abs(y_true_plage) + np.abs(y_pred_plage)).clip(lower=1e-8)) * 100


    nrmse_range = (rmse_p / range_y_mw) * 100 if range_y_mw > 0 else float('inf')
    nmse_range = (mse_p / (range_y_mw ** 2)) * 100 if range_y_mw > 0 else float('inf')

    print(f"\n Évaluation pour la plage '{plage_name}' (vent entre 3 et 25 m/s) :")
    print(f" - MSE    : {mse_p:.4f} MW²")
    print(f" - RMSE   : {rmse_p:.4f} MW")
    print(f" - MAE    : {mae_p:.4f} MW")
    print(f" - R²     : {r2_p:.4f}")
    print(f" - MAPE   : {mape_p:.4f}%")
    print(f" - SMAPE  : {smape_p:.4f}%")
    print(f" - NMSE   : {nmse_range:.4f}%")
    print(f" - NRMSE  : {nrmse_range:.4f}%")
    print(f" - NMAE   : {nmae_p:.4f}%")

# Définition des plages
mask_basse = y_test < 300
mask_moyenne = (y_test >= 300) & (y_test < 3000)
mask_haute = y_test >= 3000


eval_par_plage(y_test_mw, y_pred_final_mw, 'Basse production (<300)', mask_basse)
eval_par_plage(y_test_mw, y_pred_final_mw, 'Moyenne production (300–3000)', mask_moyenne)
eval_par_plage(y_test_mw, y_pred_final_mw, 'Haute production (>=3000)', mask_haute)
```
